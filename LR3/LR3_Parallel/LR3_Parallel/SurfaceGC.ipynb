```python
import numpy as np
import matplotlib.pyplot as plt

file = 'output_N150_T45000.txt'

df = np.loadtxt(file, delimiter=',')

#   (   -  x)
x = df[0, 1:]
#   (   -  t)
t = df[1:,0]

x, t = np.meshgrid(x, t)

#        -  u
u = df[1:, 1:]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

##ax.plot_surface(x,t,u,cmap=plt.cm.CMRmap)
##ax.plot_surface(x,t,u,cmap=plt.cm.YlGnBu_r)
ax.plot_surface(x,t,u, cmap='inferno')
##ax.plot_wireframe(x,t,u, cmap='inferno')

ax.set_xlabel("")
ax.set_ylabel("t")
ax.set_zlabel("u")

##     
angle1 = 25
angle2 = -80
ax.view_init(angle1, angle2 )
plt.savefig(file + ".png")
plt.show
```
