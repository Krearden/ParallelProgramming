```python
import numpy as np
import matplotlib.pyplot as plt

df = np.loadtxt('output_N10_T100.txt',delimiter=',')

# Первая строка (без первого значения - координаты x)
x = df[0, 1:]
# Первый столбец (без первого значения - координаты t)
t = df[1:,0]

x, t = np.meshgrid(x, t)

# Матрица без первой строки и первого столбца - значения u
u = df[1:, 1:]

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

##ax.plot_surface(x,t,u,cmap=plt.cm.CMRmap)
ax.plot_surface(x,t,u,cmap=plt.cm.YlGnBu_r)
##ax.plot_surface(x,t,u, cmap='inferno')
##ax.plot_wireframe(x,t,u, cmap='inferno')

ax.set_xlabel("х")
ax.set_ylabel("t")
ax.set_zlabel("u")

## задаем углы обзора на график
angle1 = 20
angle2 = -80
ax.view_init(angle1, angle2 )
plt.show
```
